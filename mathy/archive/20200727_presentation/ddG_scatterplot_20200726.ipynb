```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set(style="whitegrid", palette="pastel", color_codes=True, font='Helvetica')

# Set font sizes
SMALL_SIZE = 12 #6
MEDIUM_SIZE = 13 #7
BIG_SIZE = 14 #8

plt.rcParams.update({
    'legend.loc': 'center left',
    'font.family': 'Helvetica',
    'font.size': BIG_SIZE,         # controls default text sizes
    'axes.titlesize': SMALL_SIZE,  # fontsize of the axes title
    'axes.labelsize': MEDIUM_SIZE, # fontsize of the x and y labels
    'xtick.labelsize': SMALL_SIZE, # fontsize of the tick labels
    'ytick.labelsize': SMALL_SIZE, # fontsize of the tick labels
    'legend.fontsize': SMALL_SIZE, # legend fontsize
    'figure.titlesize': BIG_SIZE,   # fontsize of the figure title
    'figure.figsize': (6,6),
    'figure.dpi': 300
})
```

```python
ddg_df = pd.read_csv('../../Data/gsp1_ddg_20200726.csv', index_col=0)

# read in fitness data
dms_df = pd.read_csv('../../Data/15gen_binned_0607.csv')
dms_df.bin = dms_df.bin.mask(pd.isnull(dms_df.bin), other='drop-outs')
bin_order = ['drop-outs', 'toxic', 'STOP-like', 'intermediate', 'neutral', 'beneficial']

# fix numbering, because "description" lists mutations using re-numbered
# residues for 3m1i with the tail chopped off, rosetta numbered residues
# 1-174 correspond to Gsp1 residues 10-183
ddg_df['aa_from'] = ddg_df['mutation'].str[0]
ddg_df['position'] = ddg_df['mutation'].str[1:-1].astype('int64')
ddg_df['aa_to'] = ddg_df['mutation'].str[-1]
```

```python
df = pd.merge(dms_df, ddg_df, how='inner', on=['aa_to','position','aa_from'])
df.rename(columns={'score': 'fitness', 'total': 'ddG'}, inplace=True)
# this will only have 173 out of 174 calculated ddGs, because of Q71L mutation in the structure (so we computed L71A but the DMS scan did Q71A)

corr = df['ddg_calc'].corr(df['fitness'])

# comment this to leave NaN's out of scatterplot
df.fitness = df.fitness.fillna(min(df.fitness))
# df.fitness = df.fitness.fillna(-6)
df.bin = df.bin.apply(lambda row: 'drop-outs' if pd.isnull(row) else row)
```

```python
# prepare figure
# fig, ax = plt.subplots()

sns.scatterplot(data=df, x='fitness', y='ddg_calc',
                hue='bin',
#                 hue_order = bin_order, palette=['black']+sns.color_palette('RdBu', n_colors=7)[:5],
#                 edgecolor='black', linewidth=0.5
               )

# ax.text(-3,70,'correlation (no NA) = {}'.format(round(corr,3)))

# plt.title('Saturation mutagenesis, residues 10-183 (no tail), PDB=3m1iA')
# plt.xlabel('Fitness Score')
# plt.ylabel('ddG (REU - Rosetta Energy Units)')
# # plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

# fig.set_size_inches(8, 6)
# fig.tight_layout(rect=[0, 0, 1, 1])
# fig.savefig('satmut_ddG_2020-07-26.png', dpi=300)
```

```python
mean_df = (df
 [['aa_to','aa_from','ddg_calc']]
 .groupby(['aa_from','aa_to'])
 .agg('mean')
 .reset_index()
)

mean_df
data_to_plot = mean_df.pivot_table(values='ddg_calc', columns='aa_to', index='aa_from')
sns.heatmap(data=data_to_plot, cmap = 'bwr', center=0).figure.savefig('ddg_calc_heatmap.png')
```

```python
data_to_plot = (mean_df
                .loc[lambda df: df['aa_to'] != 'P']
                .pivot_table(values='ddg_calc', columns='aa_to', index='aa_from')
               )

sns.heatmap(data=data_to_plot, cmap = 'bwr', center=0)
```

```python
# remake scatterplot with no prolines in aa_to

df_noP = pd.merge(dms_df, ddg_df, how='inner', on=['aa_to','position','aa_from'])
df_noP.rename(columns={'score': 'fitness', 'total': 'ddG'}, inplace=True)

df_noP = df_noP[df_noP['aa_to'] != 'P']

corr = df_noP['ddg_calc'].corr(df_noP['fitness'])

# comment this to leave NaN's out of scatterplot
df_noP.fitness = df_noP.fitness.fillna(min(df.fitness))
# df.fitness = df.fitness.fillna(-6)
df_noP.bin = df_noP.bin.apply(lambda row: 'drop-outs' if pd.isnull(row) else row)

# prepare figure
fig, ax = plt.subplots()

sns.scatterplot(data=df_noP, x='fitness', y='ddg_calc', hue='bin',
                hue_order = bin_order, palette=['black']+sns.color_palette('RdBu', n_colors=7)[:5],
                edgecolor='black', linewidth=0.5)

ax.text(-3,50,'correlation (no NA) = {}'.format(round(corr,3)))

plt.title('Saturation mutagenesis, excluding "to P", residues 10-183 (no tail), PDB=3m1iA')
plt.xlabel('Fitness Score')
plt.ylabel('ddG (REU - Rosetta Energy Units)')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

fig.set_size_inches(8, 6)
fig.tight_layout(rect=[0, 0, 1, 1])
fig.savefig('satmut_ddG_noP_2020-07-26.png', dpi=300)

```
