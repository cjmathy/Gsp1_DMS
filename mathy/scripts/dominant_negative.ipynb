```r
### to get the fitness of "no protein", for calculating the dominant negative threshold
### take the average value of stop codon for the first X positions (resulting in non-functional protein)

compute_STOP_stats <- function(df) {
    df %>% 
    filter(aa_to == '*' & position <= 170) %>%
    summarize('mean' = mean(score, na.rm = T),
              'std.dev' = sd(score, na.rm = T),
              'cutoff' = mean - std.dev,
              'cutoff_3Z' = mean - 3*std.dev)
}

STOP_stats_6gen <- compute_STOP_stats(dms_6gen)
STOP_stats_15gen <- compute_STOP_stats(dms_15gen)

dms_6gen <- dms_6gen %>%
    mutate(is_dominant_negative = ifelse(
        (aa_to != '*') & ((score < STOP_stats_6gen$cutoff) | is.na(score)), T, F))

dms_15gen <- dms_15gen %>%
    mutate(is_dominant_negative = ifelse(
        (aa_to != '*') & ((score < STOP_stats_15gen$cutoff) | is.na(score)), T, F))
```

```r
# compute overlap with interfaces and functional loops

ploop <- seq(17, 23)
switchI <- seq(39, 45)
switchII <- seq(67, 75)
nuc_5A_shell <- c(seq(19, 28), seq(37, 44), seq(67,71), 124, 125, 127, 128, seq(152, 154))

nucleotide_binding_site <-
    data.frame('partner' = c(rep('P-loop', length(ploop)),
                             rep('Switch I', length(switchI)),
                             rep('Switch II', length(switchII)),
                             rep('Active site', length(nuc_5A_shell))),
               'yeastresnum' = c(ploop, switchI, switchII, nuc_5A_shell),
               stringsAsFactors=F) %>% 
    mutate('in_interface' = T)
    
# consensus residues for karyopherin binding interface
kap_consensus = c(seq(45,49), 66, 72, seq(76, 84), seq(108,115) , 142, 143, 145, 147, 157)

# also added a YRB1_patch partner, computed for the complex when the Gsp1 C-term (res 180-217)
# and Yrb1 N-term (res 71-82) are deleted from the structures
# recomputed the interface using Tina's script `calculate_interfaces_and_interface_conservation_for_all_chains.R`

interfaces <-
    read_delim('../../Data/Gsp1_interfaces_SASA_and_conservation.txt', delim='\t', col_types=cols()) %>% 
    filter(protein == 'GSP1') %>% 
    mutate(in_interface = ifelse(interface %in% c('core','rim'), T, F),
#     mutate(in_interface = ifelse(deltarASA > 0.25, T, F),
           partner = ifelse(file_path == 'PyMOL_figures/pdbs/clean/3m1i2.pdb',
                            'YRB1 patch', partner)) %>% 
    select(partner, 'yeastresnum'=resno, in_interface) %>%
    unique() %>%
    bind_rows(nucleotide_binding_site) %>%
    right_join(data.frame('yeastresnum' = seq(from = 1, to = 220)),
               by = 'yeastresnum') %>%
    complete(partner, yeastresnum) %>%
    filter(!is.na(partner)) %>%
    mutate(in_interface = ifelse(is.na(in_interface), F, in_interface)) %>% 
    bind_rows(tibble(partner = 'Kaps', yeastresnum = seq(1,220)) %>% 
              mutate(in_interface = ifelse(yeastresnum %in% kap_consensus, T, F)))
```

```r
interface_sizes <- 
    interfaces %>% 
    group_by(partner, in_interface) %>% 
    summarise(size = n(), .groups='drop') %>% 
    filter(in_interface == TRUE) %>% 
    select(-in_interface)
largest_interface_size <-
    interface_sizes %>% 
    pull(size) %>% 
    max()
```

```r
compute_jaccard <- function (df, ...) {
    
    counts <- df %>% 
      mutate(similarity = case_when(in_interface==TRUE & dom_neg_pos==TRUE ~ 'both',
                                    in_interface==TRUE | dom_neg_pos==TRUE ~ 'either',
                                    T ~ 'neither')) %>% 
      group_by(similarity) %>% 
      summarise(n = n(), .groups='keep') %>% 
      ungroup() %>% 
      deframe()
        
    if (is.na(counts['both'])) {
        counts['both'] <- 0
    }
    
    counts['jaccard'] <- counts['both']/(counts['both']+counts['either'])
    return(counts['jaccard'])
}

interface_jaccard <-
    interfaces %>% 
    mutate(dom_neg_pos = ifelse(yeastresnum %in% domneg_6gen$position, T, F)) %>% 
    group_by(partner) %>% 
    nest() %>% 
    mutate(jaccard = map_dbl(data, compute_jaccard)) %>%
    select(-data) %>% 
    arrange(desc(jaccard)) %>% 
    ungroup()
interface_jaccard
```

```r
# need to compute an expected jaccard for a theoretical interface:
#    - randomly sample n residues with replacement from positions 1-220, call those in_interface = TRUE
#    - compute 

theoretical_jaccards <- data.frame(size=integer(),
                                   jaccard_mean=double(),
                                   jaccard_sd=double())

# compute for theoretical interfaces of sizes 1-72
for (n in seq(1, largest_interface_size)) {

    # conduct i = 100 independent trials for each theoretical interface size n    
    # (I confirmed that the mean and sd of the estimate is converged after 100 trials)
    n_trials = 100
    jvec <- rep(NA, n_trials)
    for (i in seq(1, n_trials)) {
    
        theoretical_interface <- sample(219-length(active_site), n, replace = FALSE)
        
        jvec[i] <-
            data.frame('yeastresnum' = seq(from = 1, to = 219),
                       'in_interface' = rep(FALSE, 219)) %>% 
            filter(!yeastresnum %in% active_site) %>% 
            mutate(in_interface = ifelse(yeastresnum %in% theoretical_interface, T, F)) %>% 
            mutate(dom_neg_pos = ifelse(yeastresnum %in% domneg_6gen$position, T, F)) %>% 
            compute_jaccard()
    }
    
    theoretical_jaccards <- rbind(
        theoretical_jaccards,
        data.frame(size=n,
                   theoretical_jaccard_mean=mean(jvec),
                   theoretical_jaccard_sd=sd(jvec))
    )
    
}
tail(theoretical_jaccards)
```

another way to do this is to just sample from a hypergeometric distribution
- m, n are the n. of DNs and non-DNs. They add up to the total number of residues (219, note)
- k is the number of residues in your set (a theoretical interface)
- `phyper(x, m, n, k)` gives the probability distribution of finding x DNs when forming a random set of k residues
- `dhyper(x, m, n, k)` gives the CDF for finding x DNs when forming a random set of k residues

```r
# compute an independent distribution for count frequencies
# by randomly assigning the dominant negatives to other positions

dns <- domneg_6gen$position # dominant negatives

m <- length(dns)    # n. of DN positions (excludes the last codon, STOP)
t <- 219            # n. of positions not in active site
n <- t-m            # n. of tolerant not in active_site

# thus, we're computing the expectation of pulling at x dominant negatives when pulling k values
# for a given position, where k is the interface size

# expected number of positions when independently drawing
counts_null <- data.frame(list(n_dn=x, Freq=t*dhyper(x, m, n, k))) %>% 
    mutate(n_dn = factor(n_dn))


interface_counts <-
    interfaces %>% 
    mutate(dom_neg_pos = ifelse(yeastresnum %in% dns, T, F)) %>% 
    group_by(partner) %>% 
    filter(in_interface) %>% 
    summarise(n_int = sum(in_interface),
              n_dn = sum(dom_neg_pos)) %>% 
    mutate(fraction = n_dn/n_int,
           prob = 1-phyper(n_dn, m, n, n_int))
    
p1 <-
    interface_counts %>% 
    ggplot(aes(x=fraction, y=-log10(prob))) +
    geom_point(size=2) +
    geom_text_repel(aes(label=partner),size=2, vjust=1) +
    geom_hline(yintercept=-log10(0.05), linetype='dashed', size = 0.2) +
    xlab('Fraction of interface that is DN') + ylab('-log10(prob)') +
    ggtitle('All positions, probabilities from hypergeometric dist\n') +
    theme_custom + 
    theme(axis.text.x=element_text(angle=45, hjust=1, vjust=1),
         legend.key.size = unit(0.4, 'lines'))
```
