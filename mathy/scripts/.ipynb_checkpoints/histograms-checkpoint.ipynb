```python

import os
from datetime import datetime
today = datetime.now().strftime('%Y-%m-%d')


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns
from sinaplot import sinaplot



%matplotlib inline
sns.set(style="whitegrid", palette="pastel", color_codes=True, font='Helvetica')

# Set font sizes
SMALL_SIZE = 6
MEDIUM_SIZE = 7
BIG_SIZE = 8

plt.rcParams.update({
    'legend.loc': 'center left',
    'font.family': 'Helvetica',
    'font.size': BIG_SIZE,         # controls default text sizes
    'axes.titlesize': SMALL_SIZE,  # fontsize of the axes title
    'axes.labelsize': MEDIUM_SIZE, # fontsize of the x and y labels
    'xtick.labelsize': SMALL_SIZE, # fontsize of the tick labels
    'ytick.labelsize': SMALL_SIZE, # fontsize of the tick labels
    'legend.fontsize': SMALL_SIZE, # legend fontsize
    'figure.titlesize': BIG_SIZE   # fontsize of the figure title
})
```

```python
# Gsp1 sequence
WT_seq = 'MSAPAANGEVPTFKLVLVGDGGTGKTTFVKRHLTGEFEKKYIATIGVEVHPLSFYTNFGE' \
         'IKFDVWDTAGQEKFGGLRDGYYINAQCAIIMFDVTSRITYKNVPNWHRDLVRVCENIPIV' \
         'LCGNKVDVKERKVKAKTITFHRKKNLQYYDISAKSNYNFEKPFLWLARKLAGNPQLEFVA' \
         'SPALAPPEVQVDEQLMQQYQQEMEQATALPLPDEDDADL*'

# secondary structure
gsp1_ss = (
    pd.read_csv('/Users/cjmathy/gdrive/gsp1_dms/Data/gsp1_secondary_structure_wide.csv')
    [['position', 'consensus_ss']]
    .rename(columns={'aa_position': 'position'})
) 

# read in fitness data
EMPIRIC_df = pd.read_csv('/Users/cjmathy/gdrive/gsp1_dms/Data/15gen_fitness_current.csv')
EMPIRIC_df.bin = EMPIRIC_df.bin.mask(pd.isnull(EMPIRIC_df.bin), other='drop-outs')
bin_order = ['drop-outs', 'toxic', 'STOP-like', 'intermediate', 'neutral', 'beneficial']


def dict_of_lists_to_long_table(d, var_name, value_name, value_type):
    return(pd.DataFrame.from_dict(data=d, orient='index')
    .transpose()
    .melt(var_name=var_name, value_name=value_name)
    .dropna()
    .assign(position=lambda x: x[value_name].astype(value_type))
    .reindex(columns=[value_name,var_name])
    )

outdir = '../figures/python_plots/histograms_{}'.format(today)
if not os.path.exists(outdir):
    os.mkdir(outdir)
```

```python
def plot_regions_by_bins(file, df, x, order=None, title='', w=5, h=5):
    
    # prepare figure
    fig, axs = plt.subplots(1)
    fig.suptitle(title)
    
    
    frac_df = (df
        .groupby([x])
        ['bin']
        .value_counts(normalize=True, dropna=False)
        .rename('frac_in_bin')
        .reset_index()
    )
                        
    g = sns.barplot(data=frac_df, x=x, y='frac_in_bin', hue='bin',
                    order=region_order, hue_order=bin_order,
                    palette=['black']+sns.color_palette('RdBu', n_colors=7)[:5],
                    edgecolor='black', linewidth=0.5)
            
    plt.xlabel('Mutation positions')
    plt.ylabel('Fraction of \nSelection Coefficients')
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

    fig.set_size_inches(w, h)
    fig.tight_layout(rect=[0, 0, 1, 0.92])
    
    fig.savefig(file, dpi=300)
    
    return
```

```python
# conserved/switch loops
# NOTE: leave out C-term helix (194-208) bc it matches the entire protein dist,
# and it's really the C-term linker that switches conformation
loops = {
    'P-loop\n(19-25)': list(range(19, 25+1)),
    'Switch I\n(41-47)':  list(range(41, 47+1)),
    'Switch II\n(69-77)': list(range(69, 77+1)),
    'C-terminal linker\n(181-193)': list(range(181,193+1)),
}

all_loop_pos = gsp1_ss[gsp1_ss.consensus_ss == 'L']['position']
func_loop_pos = set([pos for sublist in list(loops.values()) for pos in sublist])  # just unnests the lists in the dict
loops['All other loops\n(n=76)'] = [pos for pos in all_loop_pos if pos not in func_loop_pos]
loops['Entire protein\n(1-220)'] = list(range(1, 220+1))
loops_df = dict_of_lists_to_long_table(
    d=loops, var_name='loop', value_name='position', value_type='int'
)

# merge loop annotations into dataset
loops_df = EMPIRIC_df.merge(loops_df, on='position')

# set ordering of levels in plots
region_order = ['P-loop\n(19-25)', 'Switch I\n(41-47)',
                'Switch II\n(69-77)', 'C-terminal linker\n(181-193)',
                'All other loops\n(n=76)', 'Entire protein\n(1-220)']
title = ('Conserved loops and those that change conformation upon\n'
         'GTP-binding have significantly poorer fitnesses')
w, h = 12, 4
region = 'loop'


plot_regions_by_bins(df=loops_df, x=region, title=title, w=w, h=h, order=region_order,
    file=os.path.join(outdir,'{}_plots.png'.format(region)))
```

```python
# lobes
lobes = {
    'Effector lobe\n(1-96)': list(range(1, 96+1)),
    'Allosteric lobe\n(97-180)': list(range(97,180+1)),
    'C-terminal extension\n(181-220)': list(range(181, 220+1))
}
lobes_df = dict_of_lists_to_long_table(
    d=lobes, var_name='lobe', value_name='position', value_type='int'
)

# merge lobes annotations into dataset
lobes_df = EMPIRIC_df.merge(lobes_df, on='position')

# plot settings
region_order = ['Effector lobe\n(1-96)', 'Allosteric lobe\n(97-180)', 'C-terminal extension\n(181-220)']
title = ('The allosteric lobe is more tolerant to mutation,\n'
         'and even shows an enrichment in beneficial mutations') 
w, h = 4, 2
region = 'lobe'

plot_regions_by_bins(df=lobes_df, x=region, title=title, w=w, h=h, order=region_order,
                     file=os.path.join(outdir,'{}_plots.png'.format(region)))
```

```python
# goal here is to divide all positions into categories
#    core: in the core of at least one interface, n = 70
#    rim: not ever core, but rim in at least one, n = 6
#    support: not ever in core or rim, but in support in at least one, n = 1
#    none: not in any interface

interfaces = (
    pd.read_csv('../Data/interface_residues_and_mutations_MODIFIED.txt', sep='\t')
    .rename(columns={'yeastresnum':'position'})
    [['position', 'partner', 'deltarASA','interface']]
    .drop_duplicates()
)

def label_burial(l):
    '''applies a rule to label a position by maximal burial across many structures'''

    if 'core' in l:
        return('core')
    elif 'rim' in l:
        return('rim')
    elif 'support' in l:
        return('support')
    else:
        return('none')
    
grouped = interfaces.groupby('position')['interface']
labels = [(position[0], label_burial(position[1].values)) for position in grouped]
interface_df = pd.DataFrame(data = {'position': [tup[0] for tup in labels],
                     'interface': [tup[1] for tup in labels]})
all_positions_df = pd.DataFrame(data = {'position': np.arange(1,221, dtype = int)})
interface_df = all_positions_df.merge(interface_df, how = 'outer', on='position').fillna('none')

print(interface_df.groupby('interface').count())

interface_names = {
  'core': '\n'.join(['core (70 positions)', 'rASAm > 25%', 'rASAc > 25%']),
  'rim': '\n'.join(['rim (6 positions)', 'rASAc > 25%']),
  'support': '\n'.join(['support (1 position)', 'rASAc > 0%']),
  'none': '\n'.join(['none (143 positions)', 'rASAc-rASAm = 0'])}


iface_df = (EMPIRIC_df
            .merge(interface_df)
            .replace({'interface': interface_names})
           )

# plot settings


region_order = [interface_names['core'], interface_names['rim'],
                interface_names['support'], interface_names['none']]
title = ('In gross, interface burial does predict poorer fitness') 
w, h = 5, 2
region = 'interface'

plot_regions_by_bins(df=iface_df, x=region, title=title, w=w, h=h, order=region_order,
    file=os.path.join(outdir,'{}_plots.png'.format(region)))
```

```python
# another plot we can make is to make a plot for each partner

def complete_pairs(df, column):
    return (df
     .merge(interfaces, how='outer', on = 'position')
     .pivot(index='position', columns='partner', values=column)
     .reset_index()
     .melt(id_vars='position', value_name=column)
    )
    
interface_names = {
  'core': '\n'.join(['core', 'rASAm > 25%', 'rASAc > 25%']),
  'rim': '\n'.join(['rim', 'rASAc > 25%']),
  'support': '\n'.join(['support', 'rASAc > 0%']),
  'none': '\n'.join(['none', 'rASAc-rASAm = 0'])}    

full_deltarASA = complete_pairs(df=all_positions_df, column='deltarASA')
full_interface = complete_pairs(df=all_positions_df, column='interface')

iface_df = (full_deltarASA
            .merge(full_interface, on = ['position', 'partner'])
            .dropna(how='all', subset=['partner'], axis=0)
            .assign(deltarASA = lambda df: df.deltarASA.apply(lambda row: row if pd.notnull(row) else 0))
            .assign(interface = lambda df: df.interface.apply(lambda row: row if pd.notnull(row) else 'none'))
            .replace({'interface': interface_names})
            .merge(EMPIRIC_df, on='position')
           )

# from matplotlib.backends.backend_pdf import PdfPages
# pp = PdfPages(os.path.join(outdir,'interface_partner_plots.pdf'))

for partner in iface_df.partner.drop_duplicates():
    df = iface_df[iface_df.partner == partner]
    
    region_order = [interface_names['core'], interface_names['rim'],
                interface_names['support'], interface_names['none']]
    title = partner 
    w, h = 5, 2
    region = 'interface'

    plot_regions_by_bins(df=df, x=region, title=title, w=w, h=h, order=region_order,
        file=os.path.join(outdir,'{}_plots.png'.format(partner)))
    
```
